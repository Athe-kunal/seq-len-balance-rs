```rust
use std::collections::BinaryHeap;

fn main() {
    let nums: Vec<i64> = vec![8, 7, 6, 5, 4];
    let mut heap = BinaryHeap::from(nums);

    while heap.len() > 1 {
        let a = heap.pop().unwrap();
        let b = heap.pop().unwrap();

        let diff = a - b; // no abs needed
        heap.push(diff);
    }

    println!("Final residue: {:?}", heap.pop().unwrap());
}
```
